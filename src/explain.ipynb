```python
import os
import shap
import joblib
import pandas as pd
import matplotlib.pyplot as plt

# Create plots folder if it doesn't exist
os.makedirs('../outputs/plots', exist_ok=True)

# Load model and data
pipe = joblib.load('../outputs/models/lgbm_pipeline.pkl')
df = pd.read_csv('../data/processed/fused.csv')

TARGET = 'target'
DROP_COLS = [
    c for c in ['target', 'TWF', 'HDF', 'PWF', 'OSF', 'RNF']
    if c in df.columns
]

X = df.drop(columns=DROP_COLS)

# Get LightGBM model
lgbm_model = pipe.named_steps['clf']

# Sample data for SHAP
X_sample = X.sample(500, random_state=42)

explainer = shap.TreeExplainer(lgbm_model)

# New SHAP API
explanation = explainer(X_sample)

plt.figure(figsize=(10, 7))
shap.plots.bar(explanation, show=False)

plt.tight_layout()
plt.savefig('../outputs/plots/shap_summary.png', dpi=150)
plt.close()

print("SHAP plot saved.")

plt.tight_layout()
plt.savefig('../outputs/plots/shap_summary.png', dpi=150)
plt.close()

print("SHAP plot saved.")
```
