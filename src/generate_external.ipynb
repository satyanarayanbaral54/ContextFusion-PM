```python
import pandas as pd
import numpy as np
import os

# Load data
df = pd.read_csv('../data/raw/ai4i2020.csv')
n = len(df)

# Simulate timestamps
df['timestamp'] = pd.date_range(
    start='2023-01-01',
    periods=n,
    freq='h'
)

# Simulate context signals
t = np.arange(n)

df['ambient_temp_c'] = (
    15
    + 10 * np.sin(2 * np.pi * t / (24 * 365))
    + np.random.normal(0, 2, n)
)

df['factory_load_pct'] = np.clip(
    60
    + 20 * np.sin(2 * np.pi * t / 24)
    + np.random.normal(0, 8, n),
    10,
    100
)

df['humidity_pct'] = np.clip(
    50
    + 15 * np.cos(2 * np.pi * t / (24 * 30))
    + np.random.normal(0, 5, n),
    20,
    95
)

# Create output folder if needed
os.makedirs('../data/external', exist_ok=True)

# Save file
df.to_csv('../data/external/context_signals.csv', index=False)
print("External context data saved.")
```
